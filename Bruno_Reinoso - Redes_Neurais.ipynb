```python
# ----- Biblioteca -----
import numpy as np

# Keras é uma biblioteca extensamente usada para machine learning, é uma espécie da abstração 
# que nos permite usar mais facilmente o conceito de TensorFlow, tensores (forma de generalização de matrizes,
# container de dados)

# tensor de ordem/dimensão zero --> escalar, número
# tensor de ordem um --> array, vetor
# tensor de ordem dois --> matriz


# ----- Teste escalar -----
x1 = np.array(12) # x é um número
print(x1.ndim) # vai me dar a ordem deste tensor (já embutido no numpy)

# ----- Teste array -----
x2 = np.array([0, 4, 6, 4, 8]) #x é um array
print(x2.ndim)
```

```python
# ----- Importa a biblioteca com os dados de escrita numérica (a mão) -----
from keras.datasets import mnist 
import matplotlib.pyplot

# Monta matriz com dados
(treino_im, treino_rot),(teste_im, teste_rot) = mnist.load_data()
treino_im.shape

# Teste anterior
# digito = treino_im[4]
# plt.imshow(digito, cmap = plt.cm.binary)
# plt.show()

from keras import models
from keras import layers

rede = models.Sequential()

# 512 é o número de neurônios, relu é a função de ativação, jeito da entrada da primeira camada (imagens)
# Vírgula necessária por ser apenas uma dimensão
rede.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))

# Temos 10 categorias, função de ativação 'softmax' distribui as categorias de acordo com probabilidade
rede.add(layers.Dense(10, activation='softmax'))

# otimizador, função, métrica
rede.compile( optimizer = 'rmsprop', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Deve-se normalizar os números de treino e teste
treino_im = treino_im.reshape( (60000,28 *28))
treino_im = treino_im.astype('float32') /255
teste_im = teste_im.reshape( (10000, 28 *28))
teste_im = teste_im.astype('float32') /255


# One-hot encoding --> Nos circuitos digitais e no aprendizado de máquina, o one-hot é um grupo de bits entre 
# os quais as combinações legais de valores são apenas aquelas com um único bit alto e todas as outras baixas

from keras.utils import to_categorical

# One-hot encoding
treino_rot = to_categorical(treino_rot)
teste_rot = to_categorical(teste_rot)

# ----- Ajuste fit -----

# Dados de treino, vezes que ele vai treinar, quanto vai usar para treinar (número amostras)
print('----- Treino da Rede Neural -----')
rede.fit( treino_im, treino_rot, epochs=30, batch_size=128)

# Possível overfit --> É um termo usado em estatística para descrever quando um modelo 
# estatístico se ajusta muito bem ao conjunto de dados anteriormente observado, mas se mostra ineficaz para prever novos resultados.

# Desempenho da rede
print('\n----- Teste da Rede Neural -----')
erro_teste, acu_teste = rede.evaluate(teste_im,teste_rot)
# Substituindo epochs de 5 para 30 conseguimos uma acurácia maior com batch_size=128. 
# Além desse valor escolhido para epochs a acurácia não necessariamente aumenta
```
